### Overview : we are going to give ranks to players on the basis of salary and then sort them on the basis of salary rank . How cooool

## Alternative way ::  sort_values["Salary"]

```python
import pandas as pd
```

```python
nba=pd.read_csv("nba.csv")
nba

```

```python
nba.dropna(how="all" , inplace =True)
nba
```

```python
nba["Salary"]=nba["Salary"].fillna(0).astype("int")
nba["Salary"]
```

```python
nba
```

```python
nba["salary rank"]=nba["Salary"].rank(ascending=False).astype("int")
nba["salary rank"]
```

```python
nba
```

```python
nba.sort_values("Salary")
```

```python
nba.sort_values("salary rank")
```
