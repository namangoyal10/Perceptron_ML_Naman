# Efficient method to save memory
 - astype("category")
 it assigns the row the category according the different categories available in the column
 ex : POSITION has 5 categories : pg,sf,c,etc. Rather than storing a position object for each row it just stores one of the five categories it belongs to

```python
nba.info()
```

```python
nba["Position"].nunique()
nba["Position"]=nba["Position"].astype("category")
nba
```
