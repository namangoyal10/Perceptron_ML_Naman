# Imp concept

### Axis = 0 : traversing in every column of a particular row ( operation performed is on the row)
### Axis = 1 : traversing in every row of a particular column ( operation performed is on the column)

### "how" parameter decides the behaviour of removal
- all means removes the rows with all values null
- any mean removes the rows having any (even a single value as null )

## subset = [ ] accepts a list of column name in which we want to look for null values exclusively

```python
import pandas as pd
```

```python
nba1=pd.read_csv("nba.csv")
nba2=pd.read_csv("nba.csv")
nba3=pd.read_csv("nba.csv")

nba1.tail()
### last column is null as shown :
```

```python
nba1.dropna(axis = 0 ,inplace = True) 
nba1.tail()
```

```python
nba2.dropna( inplace = True ,how="all") 
print(len(nba2))
nba2.tail()
```

```python
nba3.dropna(axis =1 , inplace=True , how = "any")
nba3
```

```python
nba1.dropna(axis= "rows" ,how="any")
nba1
```

```python
rev=pd.read_csv("revenue.csv")
rev
```

```python
rev.dropna(axis=0 , how="any" )
```

```python
rev.dropna(axis=0 , how="all" )
```

```python
rev.dropna(axis=0 , how="all" , subset=["Drop Test" , "Drop test1"] )
## the keyword "how" has no meaning if we give only one parameter to the subset (think)
```

```python
rev.dropna(axis=1 , how="all" )
```
