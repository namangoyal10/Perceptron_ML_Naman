# By Index Label

```python
import numpy as np 
import pandas as pd
```

```python
pokemon= pd.read_csv("pokemon_full.csv")
pokemon
```

```python
pokemon=pd.read_csv("Pokemon_full.csv" , index_col=["Pokemon"] )
pokemon
```

```python
pokemon= pd.read_csv("pokemon_full.csv", index_col="Pokemon" , usecols=[ "Pokemon", "Type 1"], squeeze= True)
pokemon.head(3)
```

# THE INDEX HAS NOW GOT THE LABELS OF POKEMON NAMES AND NOW CAN BE USED TO EXTRACT THE SERIES VALUE . 

# ALSO THE VALUES STILL HAVE INDEX POSITIONS ALLOCATED IN MEMORY WHICH IS THOUGH NOT VISIBLE BUT STILL USABLE

```python




print(pokemon[0])
print(pokemon.index[0])
print(pokemon.values[0])
print(pokemon["Hoopa"])
```

```python


pokemon.sort_index(ascending=True)
```

```python


pokemon.sort_values(ascending=True)
```
