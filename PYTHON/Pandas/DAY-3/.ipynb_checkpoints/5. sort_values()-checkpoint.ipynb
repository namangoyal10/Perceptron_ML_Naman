```python
import pandas as pd
```

```python
nba=pd.read_csv("nba.csv")
nba
```

# Type 1 : Single value as a parameter to sort 

```python
nba.sort_values("Salary" , ascending=False , inplace=True , na_position="first")
## na_pos means where should be the null values placed at first or at last
nba
```

# Type 2: Multiple values chained as a parameter

## First sort according to team name and then according to player name

```python
nba1=pd.read_csv("nba.csv")
nba1.sort_values(["Team" , "Name"] , ascending=True , inplace=True , na_position="first")
nba1
```

## We can give ascending for both the parameters differently as well

```python
nba2=pd.read_csv("nba.csv")
nba2.sort_values(["Team" , "Name"] , ascending=[True ,False] , inplace=True , na_position="first")
nba2.head(20)
```
