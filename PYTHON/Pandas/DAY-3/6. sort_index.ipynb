```python
import pandas as pd
```

## Example one : 

```python
nba=pd.read_csv("nba.csv")
nba.sort_values("Weight" ,inplace=True)
nba
```

```python
nba.sort_index(ascending=False , inplace=True , na_position="first")
nba
```
