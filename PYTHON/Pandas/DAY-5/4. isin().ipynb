# Aim :  to check multiple conditions in a single column itself

```python
import pandas as pd
```

```python
emp = pd.read_csv("employees.csv")
emp
```

# Method -1 : Bruteforce way

```python
cond1 = emp[ "Team" ] == "Finance"
cond2 = emp[ "Team" ] == "Engineering"
emp[ cond1 | cond2 ]
```

# Method - 2 : isin()

```python
cond3 = emp["Team"].isin( ["Finance", "Engineering" ] ) 
emp= emp[cond3]
emp
```
