```python
import pandas as pd
```

```python
emp=pd.read_csv("employees.csv")
```

```python
emp
```



# AND & conditions


```python


job_one = emp["Salary"] > 70000
job_two = emp["Bonus %"]>7.00
```

```python
emp[job_one & job_two]

## here emp[job_one and job_two] doesnt work
```

# OR | conditions

```python
emp[job_one | job_two]

## here emp[job_one and job_two] doesnt work
```

# Some more example

```python
job_three = emp["Gender"] == "Female"

emp[job_three & (job_one | job_two) ]

# brackets are a must . idk what happens if we dont use them
```

```python
job_four = emp["Team"] == "Finance" 
job_five= emp["Team"] == "Engineering"
emp[ job_four | job_five ]

```
