# Aim :  to extract the values which are between a range 
## Returns a boolean Series

Note  :  to access date and time for .between() they must be coverted to_datetime() or parse dates_first

```python

import pandas as pd
```

```python
emp = pd.read_csv("employees.csv" , parse_dates= ["Last Login Time"])
emp
```

```python
emp["Salary"].between( 10000 , 90000)
```

```python
condition_one = emp["Salary"].between( 10000 , 90000)
emp[condition_one]
```

```python
condition_two = emp["Bonus %"].between(4,7)
emp[condition_two]
```

```python
emp["Start Date"]  =  pd.to_datetime(emp["Start Date"])
emp["Start Date"].between("1995-01-15" , "1999-01-20")
```

```python
condition_three = emp["Start Date"].between("1995-01-15" , "1999-01-20")
emp[condition_three]
```

```python
emp[emp["Last Login Time"].between("07:30AM" , "7:30PM" )]
```
