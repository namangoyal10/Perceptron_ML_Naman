# Filtering dataframe based on a condition



```python
import pandas as pd
```

# Method - 1 

```python
emp=pd.read_csv("employees.csv" , parse_dates=["Start Date" , "Last Login Time"]) 
emp
```

```python
emp["Gender"]
```

```python
## accessing only male
emp["Gender"] == "Male"
```

```python
emp[ emp["Gender"] == "Male"] 
```

# Method - 2


```python
var =( emp["Team"] == "Finance" )
print(var)
```

```python
emp [ var ]
```

## if a column is already of boolean data type - Doesnot work on columns with missing values

```python

emp.dropna(how="any" , inplace=True)
emp
```

```python
emp[ emp["Senior Management"] ]
```

## Trick :  emp[ emp["Senior Management"] == True ]
## This doesnt care about null values 

## Some more examples

```python
emp[ emp["Team"] != "Marketing" ]
```

```python
emp[ emp["Bonus %"] > 5 ]
```

```python
emp[ emp["Start Date"] <= "1999-02-10" ]
```
