```python
a= np.array([[1,2,3,4],[5,6,7,8]])
```

```python
print(np.min(a))
print(np.min(a ,axis =0 )) ## row is changing while the column remains the same
print(np.min(a ,axis =1 ))

print(np.max(a))
print(np.max(a ,axis =0))
print(np.max(a, axis =1))
```

```python
#mean
b=np.array([1,2,3,4,5])
c=np.array([[1,2],[3,4]])
m1 = np.sum(b) / len(b)
print(m1)

m2=np.mean(b)
print(m2)

m3=np.mean(c) ## mean of all the elements
print(m3)

m4=np.mean(c , axis=0)
print(type(m4),m4)
```

```python
print(np.median(b))
print(np.median(c))
print(np.median(c ,axis=1))

##confusion  are mean and median always the same = NO
d=np.array([1,2,3,0,5])
print(np.mean(d),np.median(d))


```

```python
## average = weighted mean
## mean = 1+2+3+4+5 /5
## average= (1*1 + 2*3 + 3*7 +4 *9) /4   weights are manually assigned and multiplied

n=np.array([1,2,3,4,5])
w=np.array([1,10,20,15,4])

print(np.mean(n))
#print(np.average(n , weights=w))   
print(np.average(n,weights=w))
```

# STANDARD DEV (sigma) and VARIANCE ( SD^2)



```python
p=np.array([1,4,3,6,2])
u=np.mean(p)
myStd=np.sqrt(np.mean(abs(p-u)**2))
npStd=np.std(p)
print(myStd,npStd)
```

```python
print(np.var(p))
```

```python
print(np.std(a,axis=0) , np.std(a,axis=1)  ,sep = "\t")
print(np.var(a,axis=0) , np.var(a,axis=1)  ,sep="\t")
```
