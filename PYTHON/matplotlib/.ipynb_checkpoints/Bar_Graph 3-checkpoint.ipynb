```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use("seaborn")
```

```python
plt.xlabel("TURN")
plt.ylabel("SCORE")


a=np.array([1,2,3,4])
b=np.array([5,10,15,20])
##default align=centre
a1=a+0.2
b1=np.array([12,3,8,10])

plt.xlim(0,max(a)+1)
plt.bar(a,b  ,width=0.2 , align='edge',label="P1" , tick_label=["turn1","turn2","turn3","turn4"])
plt.bar(a1,b1  ,width=0.2 ,align='edge',label="P2" )
plt.legend()
plt.show()
```

```python
b=np.array([5,10,15,20])
b1=np.array([12,3,8,10])
a=np.array([1,2,3,4])

plt.bar(a,b ,color="green")
plt.bar(a,b1 ,color="brown")
plt.show()

plt.bar(a,b1,color="blue",alpha=0.9)
plt.bar(a,b,color="green",alpha=0.7)
plt.show()
```
