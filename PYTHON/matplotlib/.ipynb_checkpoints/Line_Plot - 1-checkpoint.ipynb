```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.use("ggplot")
```

```python
x=np.arange(1,10)
y1=x*3
y2=x**2
```

```python
plt.plot(x,y1,color="brown" , label="Dubai",linestyle="dashed",marker="^")
plt.plot(x,y2,color="green" ,label="India" , marker = "o")
plt.legend() ##to show the label of red and green curves
plt.title("Petrol")
plt.xlabel("Time")
plt.ylabel("Price")
plt.show()
```
