```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use("seaborn")
```

```python
x=np.arange(1,10)
y1=x*3
y2=x**2
```

```python
plt.figure(figsize=(20,10))
plt.scatter(x,y1,color="brown" , label="Dubai",marker="^")
plt.scatter(x,y2,color="green" ,label="India", marker="*")
plt.legend() ##to show the label of red and green curves
plt.title("Scatter Plot for Petrol")
plt.xlabel("Time")
plt.ylabel("Price")
plt.show()
```
