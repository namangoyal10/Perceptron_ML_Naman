```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use("seaborn")
```

```python
w = [ 20,10,15,5]
subject = list(["english","chemistry" , "physics" , "maths"])
print(subject,type(subject))
```

```python
plt.pie( w , labels= subject ,explode= (0.0 ,0.1 ,0.5, 0.0) , autopct='%1.1f%%',shadow=True)
plt.show()

plt.pie( w , labels= subject ,explode= (0.0 ,0.1 ,0.5, 0.0) , autopct='%1.1f%%',shadow=True ,counterclock=False)
plt.show()
```
