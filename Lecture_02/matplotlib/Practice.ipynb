```python
x=np.array([1,2,3,4,5])
y1=x**2
y2=x*3
plt.bar(x,y1,alpha=1)
plt.bar(x,y2,alpha=0.6 ,color="yellow")
plt.show()
```

```python
x=np.arange(1,6)
y1=[2,5,1,3,6]
y2=[4,1,5,2,8]


plt.bar(x-0.2,y2 , label="dhoni",alpha=1 ,align="edge",width=0.4)
plt.bar(x+0.2,y1,color="yellow", label="kohli",alpha=0.6,align="edge",width=0.4)
plt.legend()
plt.xlabel("Matches")
plt.ylabel("Sixes")
plt.ylim(0,(max(max(y1),max(y2)))+2)
plt.show()



```
