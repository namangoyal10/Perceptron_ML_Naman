```python
import matplotlib.pyplot as plt
import numpy as np
import math
plt.style.use("seaborn")
```

# ## Xsn =(x - u)/sigma

```python
Xsn=np.random.randn(100)
```

```python
plt.hist(Xsn)
plt.show()
```

# This is a standard normal distribution Histogram where sigma=1 and u=0
## u = all the values are concentrated about mean . Responsible for shift in axis
## sigma - deviation from mean ( Higher sigma - more deviat. - more expanded about mean

```python
u=50
sigma=5

## Xsn =(x - u)/sigma

x=(Xsn*sigma) + u
plt.hist(x)
plt.show()
```

```python
plt.hist(x)
plt.hist(Xsn)
plt.show()
```
